```python
df = pd.read_csv("Diwali Sales Data.csv",encoding="unicode_escape")
```

```python
df.drop(columns=['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
# Gender wise count
ax=sns.countplot(x='Gender',data=df,hue="Gender")
ax.containers
for bars in ax.containers:
 ax.bar_label(bars)
```

```python
# Gender wise total amount.
group_by=df.groupby(["Gender"],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=group_by,x='Gender',y='Amount',hue="Gender")
```

```python
# find  total count each age bin based on gender.
age=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in age.containers:
 age.bar_label(bars)
```

```python
#total amount vs age_group
age1= df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values (by='Amount',ascending=False)
sns.barplot(data=age1,x='Age Group',y='Amount')
```

```python
# total number of orders from the top 10 states.
orders= df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data=orders,x='State',y='Orders',hue="State")
```

```python
# total amount vs top 10 States.
orders= df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data=orders,x='State',y='Amount',hue="State")
```

```python
# count of marital status
ax= sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(2,5)})
for bar in ax.containers:
  ax.bar_label(bar)
```

```python
ax2= df.groupby(["Marital_Status",'Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(7,5)})
a=sns.barplot(data=ax2,x='Marital_Status',y='Amount',hue='Gender')
#for bars in ax2.containers:
#ax2.bar_label(bars)
```

```python
# finds the diff sectors
sns.set(rc={'figure.figsize':(18,5)})
occ =sns.countplot(data=df,x='Occupation',hue='Occupation')
for bars in occ.containers:
  occ.bar_label(bars)
```

```python
#Amount vs Occupation
sns.set(rc={'figure.figsize':(20,5)})
occu=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False).head(10)
sns.barplot(data=occu,x='Occupation',y='Amount',hue='Occupation')
```

```python
# count the no.of products
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# amount vs product
sns.set(rc={'figure.figsize':(20,5)})
pro_cnt=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(data=pro_cnt,x='Product_Category',y='Amount',hue='Product_Category')
```

```python
#find the total orders done by each product_id.
sns.set(rc={'figure.figsize':(15,5)})
pro_orders=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
ax=sns.barplot(data=pro_orders,x='Product_ID',y='Orders')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#total amount vs product_id
sns.set(rc={'figure.figsize':(15,5)})
pro_orders=df.groupby(['Product_ID'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
ax=sns.barplot(data=pro_orders,x='Product_ID',y='Amount', hue='Product_ID')
for bars in ax.containers:
  ax.bar_label(bars)
```
